```python
sliced_frame = DataFrame[['iso_country','type']]
by_country_type = sliced_frame.groupby(['iso_country', 'type'])
#by_country_type
agg_counts = by_country_type.size().unstack().fillna(0)
agg_counts.to_csv('airport_type_per_country_counts.csv')
agg_counts
```

```python
#use sort in ascending order
indexer = agg_counts.sum(1).argsort()
count_subset = agg_counts.take(indexer)[-15:]
count_subset.plot(kind='barh', stacked=True)
plt.savefig('counts_type_country.png')
```

```python
normed_subset = count_subset.div(count_subset.sum(1),axis=0)
normed_subset.plot(kind='barh', stacked=True)
plt.title("Airport type distribution for the top 15 African countries")
plt.savefig('normalised_counts_type_country.png')
```
